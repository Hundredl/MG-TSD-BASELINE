# 计算结果

```python
# 读入文件，在./wandb_export_2023-11-17T09_02_42.793+08_00.csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 读入文件
df = pd.read_csv('/home/v-wuyueying/workspace/blob/drive/mgtsd/baseline/patchtst/results/metrics_TimeDiff_all.csv')
```

```python
print(df.shape)
df.head()
```

```python
dataset_names = ['elec','sol','cup','traf','taxi','wiki']
results = {'dataset':[], 'metric_name':[],'value_str':[]}
for dataset_name in  dataset_names:
    # filter the df
    df_dataset = df[df['dataset'] == dataset_name]
    # 筛选出multisample_glu_rs开头的列
    cols = list(df_dataset.filter(regex='multisample_glu_rs').columns)
    for col in cols:
        # 计算multisample_glu_rs开头的列的均值和方差，保留4位小数 mean±std
        df_dataset_mean = df_dataset[col].mean()
        df_dataset_std = df_dataset[col].std()
        # 保存到文件
        results['dataset'].append(dataset_name)
        results['metric_name'].append(col)
        results['value_str'].append(f'{df_dataset_mean:.4f}±{df_dataset_std:.4f}')
# 保存到文件
df_results = pd.DataFrame(results)
df_results.to_csv('./metrics_TimeDiff_all.csv', index=False)
    
```

# 统计rank

```python
# result 读入 文件 ./result_all.xlsx
# 读入文件
import pandas as pd
df = pd.read_excel('./result_all.xlsx')
# 如果值是 - ,则替换成100，如果值是其他字符串，则替换成前6个字符所表示的数字
df = df.replace('-','9.0000±0.0000')
# 对于列 Solar	Electricity	Traffic	KDD-cup	Taxi	Wikipedia， ±前的字符串转换为数字
dataset_names = ['Solar','Electricity','Traffic','KDD-cup','Taxi','Wikipedia']
for dataset_name in dataset_names:
    df[dataset_name] = df[dataset_name].apply(lambda x: float(x.split('±')[0]))
df
```

```python
# 按Metric 分组，统计每个 Method 在各个数据集上rank的平均值
df_group = df.groupby('Metric')
df_rank = []
for name, group in df_group:
    print(name)
    cur_df = group.copy()
    # 对于每个数据集，计算rank
    for dataset_name in dataset_names:
        cur_df[dataset_name] = cur_df[dataset_name].rank()
    # 计算每个Method在各个数据集上rank的平均值
    cur_df['rank_avg'] = cur_df[dataset_names].mean(axis=1).round(1)
    # print(cur_df)
    # 保存到df_rank
    df_rank.append(cur_df)
df_rank = pd.concat(df_rank)
# 保留1位小数，输出到文件，result_all_rank.xlsx
df_rank = df_rank.round(1)
df_rank['rank_avg'] = [f'{x:.1f}' for x in df_rank['rank_avg']]
df_rank.to_excel('./result_all_rank.xlsx', index=False)
df_rank['rank_avg'] = df_rank['rank_avg'].astype(float)
df_rank
```

```python
# df_rank 按照方法名分组，对每个数据集，分别计算rank的平均值
df_rank_dataset = df_rank.copy()
df_rank_group = df_rank_dataset.groupby('Method')
df_rank_avg_dataset = []
for name, group in df_rank_group:
    cur_df = group.copy()
    # df_rank_avg_dataset.loc[len(df_rank_avg_dataset)] = [name] + list(cur_df[dataset_names].mean(axis=0)) + ['avg_rank', cur_df['rank_avg'].mean()]
    # 添加一行，Method为name，Metric为rank_avg，其他列为各个数据集上rank的平均值
    if len(cur_df) == 2:
        # df_rank_dataset.loc[len(df_rank_dataset)] = [name] + ['-' for _ in dataset_names] + ['avg_rank', 100]
        pass
    else:
        # 保留1位小数
        df_rank_dataset.loc[len(df_rank_avg_dataset)] = [name] + list(cur_df[dataset_names].mean(axis=0).round(1)) + ['avg_rank', cur_df['rank_avg'].mean().round(1)]
        df_rank_dataset.loc[len(df_rank_dataset)] = [name] + list(cur_df[dataset_names].mean(axis=0).round(1)) + ['avg_rank', cur_df['rank_avg'].mean().round(1)]
# 保留1位小数，输出到文件，result_all_rank_dataset.xlsx
df_rank_dataset = df_rank_dataset.round(1)
df_rank_dataset[dataset_names] = df_rank_dataset[dataset_names].round(1)
# 只筛选出metric为avg_rank的行
df_rank_dataset = df_rank_dataset[df_rank_dataset['Metric'] == 'avg_rank']
# rank_avg重新计算，为数据集上rank的平均值
df_rank_dataset['rank_avg'] = df_rank_dataset[dataset_names].mean(axis=1).round(1)
# 排序
df_rank_dataset = df_rank_dataset.sort_values(by='rank_avg')
df_rank_dataset.to_excel('./result_all_rank_dataset.xlsx', index=False)
df_rank_dataset
```

```python
# 计算每个metric上，MG-TSD比初次之外的最好的方法好的百分比
df_group = df.groupby('Metric')

upgrad_percent = []
for name, group in df_group:
    print(name)
    cur_df = group.copy()
    # 拿到MG-TSD的行
    mg_tsd_row = cur_df[cur_df['Method'] == 'MG-TSD']
    cur_df = cur_df[cur_df['Method'] != 'MG-TSD']
    cur_df = cur_df[cur_df['Method'] != 'MG-Input']
    # cur_df = cur_df[cur_df['Method'] != 'GP-Copula']
    # 计算每个数据集上，的min
    cur_df.loc[len(cur_df)] = ['min'] + list(cur_df[dataset_names].min(axis=0)) + ['min']
    # 计算减小的百分比
    min_row = cur_df[cur_df['Method'] == 'min']
    percents = (mg_tsd_row[dataset_names].values - min_row[dataset_names].values) / min_row[dataset_names].values
    # 转换为百分比
    percents = [f'{-x*100:.1f}%' for x in percents[0]]
    print(percents)
    # cur_df.loc[len(cur_df)] = ['percent'] + list((mg_tsd_row[dataset_names].values - min_row[dataset_names].values) / min_row[dataset_names].values) + ['percent']
    # cur_df.loc[len(cur_df)] = ['percent'] + list((mg_tsd_row[dataset_names].values - cur_df[dataset_names].values) / cur_df[dataset_names].values) + ['percent']
    # print(cur_df)
    upgrad_percent.append([name] + percents)
upgrad_percent = pd.DataFrame(upgrad_percent, columns=['Metric'] + dataset_names)

# 保存到文件
upgrad_percent.to_excel('./upgrad_percent.xlsx', index=False)
upgrad_percent
```
